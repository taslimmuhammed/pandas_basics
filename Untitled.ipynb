# Dealing with null values

```python
df = pd.read_csv(r"C:\Users\tasli\Desktop\Folders\AI\pandas\sample.csv")
print(df.isnull().sum())
```

droping null values

```python
df2 = df.dropna() #drop all the null values
print(df.shape,df2.shape)
```

if we set axis=0, then it'll remove rows, if we set the axis to 1 then it'll remove columns

```python
df3 = df.dropna(axis=1)
print(df3.shape)
```

"how"

```python
df.dropna(how="any")#remove row if any cell have null values
df.dropna(how="any")#remove row if all cell have null values
```
